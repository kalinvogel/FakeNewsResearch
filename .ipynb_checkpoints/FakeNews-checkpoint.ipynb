```python
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r"[\w']+")
df["tokenizedSentence"]=df["Text"].map(tokenizer.tokenize)

```

```python
from nltk.stem import PorterStemmer

PorterStemmer=PorterStemmer()
df["PorterStemmer"] = df["tokenizedSentence"].apply(lambda x : [PorterStemmer.stem(word) for word in x])
```

```python
from nltk.stem import LancasterStemmer

LancasterStemmer=LancasterStemmer()
df["LancasterStemmer"]=df["tokenizedSentence"].apply(lambda x : [LancasterStemmer.stem(word) for word in x])
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer=WordNetLemmatizer()
df["Lemmatizer"]=df["tokenizedSentence"].apply(lambda x : [lemmatizer.lemmatize(word) for word in x])
```

```python
df["PorterStemmerString"]=df["PorterStemmer"].apply(lambda x:' '.join(x))
```

```python
df["tokenizedSentenceString"]=df["tokenizedSentence"].apply(lambda x:' '.join(x))
```

```python

df.head()
```

```python
from sklearn.model_selection import train_test_split


X_train,X_test, y_train, y_test = train_test_split(df['PorterStemmerString'],df['Labels'],test_size=.2,random_state=7)
#X_train,X_test, y_train, y_test = train_test_split(df['tokenizedSentenceString'],df['Labels'],test_size=.2,random_state=7)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
all_text = pd.concat([X_train,X_test]) 
tfidf_vectorizer = TfidfVectorizer()
# Fit the vectorizer on the entire dataset

X_train_tfidf=tfidf_vectorizer.fit_transform(X_train)

#Transform training and test data using the same vectorizer
#X_train_tfidf = tfidf_vectorizer.transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train_tfidf, y_train)
```

```python
# Before SMOTE
print("Class distribution before SMOTE:")
print(df['Labels'].value_counts())

# After SMOTE
print("\nClass distribution after SMOTE:")
print(pd.Series(y_train_smote).value_counts())
```

```python
import xgboost as xgb

estimator_xgb = xgb.XGBClassifier(base_score=0.5, booster='gbtree', colsample_bylevel=1,
              colsample_bynode=1, colsample_bytree=1, gamma=0, gpu_id=-1,
              importance_type='gain', interaction_constraints='',
              learning_rate=0.300000012, max_delta_step=0, max_depth=6,
              min_child_weight=1, monotone_constraints='()',
              n_estimators=100, n_jobs=0, num_parallel_tree=1,
              objective='multi:softprob', random_state=0, reg_alpha=0,
              reg_lambda=1, scale_pos_weight=None, subsample=1,
              tree_method='exact', validate_parameters=1, verbosity=None)
#scale_pos_weight=scale_pos_weights,
#estimator_xgb = xgb.XGBClassifier()
estimator_xgb.fit(X_train_smote, y_train_smote)
```

```python

predictions_xgb = estimator_xgb.predict(X_test_tfidf) #prediction on train set
```

```python
from catboost import CatBoostClassifier

cat_model = CatBoostClassifier(learning_rate=.1,depth=6, l2_leaf_reg=3,iterations=100)
cat_model.fit(X_train_tfidf,y_train)
predictions_catboost = cat_model.predict(X_test_tfidf)
```

```python
from sklearn.metrics import confusion_matrix
confusion_mat = confusion_matrix(y_test, predictions_xgb)
print("Confusion Matrix:")
print(confusion_mat)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#total_instances_per_class = confusion_mat.sum(axis=1)  # Calculate total instances per true class
#percentage_confusion_mat = confusion_mat / total_instances_per_class[:, None]  # Calculate percentages
confusion_df = pd.DataFrame(confusion_mat, index=[0, 1, 2, 3, 4, 5], columns=[0, 1, 2, 3, 4, 5])

sns.heatmap(confusion_mat, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('XGBoost Smote 6-Class HeatMap')
plt.savefig('confusion_matrix_heatmap.png')
plt.show()

excel_filename = 'classification_report.xlsx'
confusion_df.to_excel(excel_filename)
```

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy = accuracy_score(predictions_xgb, y_test)
report = classification_report(predictions_xgb, y_test, output_dict=True)

df_reportxgb = pd.DataFrame(report)

excel_filename = 'classification_report.xlsx'
df_reportxgb.to_excel(excel_filename)

print('XGBoost')
print(f'Accuracy: {accuracy}')
print(f'Classification Report:\n{report}')
```

```python
accuracy = accuracy_score(predictions_catboost, y_test)
report = classification_report(predictions_catboost, y_test, output_dict=True)

df_reportcat = pd.DataFrame(report)

excel_filename = 'classification_report.xlsx'
df_reportcat.to_excel(excel_filename)

print(f'Catboost\nAccuracy: {accuracy}')
print(f'Classification Report:\n{report}')
```
